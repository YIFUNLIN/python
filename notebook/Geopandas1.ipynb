```python
import pandas as pd 
df=pd.read_csv('../geodata/Library/Library.csv',encoding='utf-8')
df
```

```python
import requests,json
def rest(address):
    # YOUR_KEY is needed
    # 範例API KAY: Bh5VwmNLww9G0Hv7sM3HnfOWL
    url = 'https://api.tomtom.com/search/2/geocode/{}.json?&key=OmECZGPSHwzh6v2LmV3RwlVbGoitlWHT&countrySet=TWN&language=zh-TW&limit=1'.format(address)
    response = requests.get(url)
    data = response.text
    js = json.loads(str(data))
    return [js['results'][0]['position']['lon'],js['results'][0]['position']['lat']]
```

```python
rest(df['address'][0])
```

```python
import geopandas as gpd
from shapely.geometry import Point
geom=[Point(rest(row.address)) for idx, row in df.head(20).iterrows()]
crs = {'init': 'epsg:4326'} #4326, 3826
gdf2 = gpd.GeoDataFrame(df.head(20), crs=crs, geometry=geom)
gdf2
```

## intersection

```python
village=gpd.read_file('../geodata/Village/village.shp',encoding='utf-8')
village.crs= {'init' :'epsg:3826'} # 避免資料沒設，這邊再重新給一次
village.head(3)
```

```python
lib=gpd.read_file('../geodata/Library/Library.shp',encoding='utf-8')
lib.crs= {'init' :'epsg:4326'} # 避免資料沒設，這邊再重新給一次
lib=lib.to_crs(epsg=3826)
lib.head(3)
```

```python
lib[lib['address'].str.contains('中和區')]
```

```python
base=village[village['ADMIT']=='中和區'].plot(color='yellow')
lib[lib['address'].str.contains('中和區')].plot(ax=base)
```

```python
import numpy as np
counts=[np.sum(row['geometry'].intersects(lib.unary_union)) for idx, row in village.iterrows()]
village['count']=counts
aggre_v=village[['ADMIV','ADMIT','count','geometry']]
aggre_v[aggre_v['count']>0]
```

```python
aggre_v.plot('count')
```

```python
dis=aggre_v.dissolve('ADMIT',aggfunc='sum')
dis
```

```python
dis.plot('count')
```
