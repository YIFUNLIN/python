```python
import folium
import pandas as pd
import wget           #觸發超連結並下載

url = 'https://tcgbusfs.blob.core.windows.net/dotapp/youbike/v2/youbike_immediate.json'

wget.download(url,'haha.json')   #引數argument

df = pd.read_json('haha.json')   #將json轉成表格模式

df.to_csv(r'C:\Users\yifun\Desktop\python\a.csv',encoding='utf-8')
```

```python
df
```

```python
bike_map = folium.Map(location=[25.03746,121.564558],     #佈置地圖(建構一個底圖)  這邊先用台北市的經緯度當畫布
                     zoom_start = 10,width = '70%',height = '70%')
bike_map
```

```python
converted_data = pd.read_csv('a.csv')

draw_map = folium.map.FeatureGroup()   #想繪製地圖內的元素(開啟繪製的權限)

for lat,lng,sbi,bemp in zip(converted_data.lat,converted_data.lng,   #有四個對象，將四個欄位zip()起來，縫在一起才可以動態拜訪
                           converted_data.sbi,converted_data.bemp):
    #lag/lng經緯度、sbi可取車輛數、bemp可停車格數
    
    if sbi == 0:                           #若沒有車可以取
        draw_map.add_child(   #針對draw_map添加資料點，畫到資料布上
        
        folium.CircleMarker([lat,lng],radius=4,color='blue',fill = True,fill_color = 'blue',fill_opacity=0.5)  
                             #經緯度    #半徑                                                  #透明度
        )
    
    if bemp == 0:                           #若沒有車可以取
        draw_map.add_child(   #針對draw_map添加資料點，畫到資料布上
        
        folium.CircleMarker([lat,lng],radius=4,color='red',fill = True,fill_color = 'red',fill_opacity=0.5)  
                             #經緯度    #半徑                                                  #透明度
        )
    
#以上所整理好的底圖內部資訊後，都放在draw_map裡    

bike_map.add_child(draw_map)    #將布置好的地圖與資料點一併呈現
                                #利用 .add_child()  將底圖內的資訊添加到初始畫布(bike_map)中
```

# 用熱力圖呈現

```python
from folium.plugins import HeatMap

san_map1 = folium.Map(location=[25.03746,121.564558],zoom_start = 12,width = '50%')    #準備台北市的畫布

df2 = df[(df.sbi == 0)]    #目前沒有車，全部為0的站台
heatdata = df2[['lat','lng']].values.tolist()
HeatMap(heatdata).add_to(san_map1)

san_map2 = folium.Map(location=[25.03746,121.564558],zoom_start = 12,width = '50%')    #準備台北市的畫布

df3 = df[(df.bemp == 0)]    #目前沒有車，全部為0的站台
heatdata2 = df3[['lat','lng']].values.tolist()
HeatMap(heatdata2).add_to(san_map2)
```

```python
san_map1.save('map.html')
```

```python
san_map2.save('map2.html')
```
